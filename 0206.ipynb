```python
#analysis : 분석
#sentiment : 감성
#classifier : (학급/반)-er (하는 주체): 학급을 만드는 주체: 분류기
#감정 ( 긍정적/부정적)
classifier = pipeline('sentiment-analysis')
# list vs string 
results=classifier(["it's very good day.", "i am sad."])
#반복
for result in results:
  print(f"label:{result['label']},with score:{result['score']}")
```

```python
from transformers import AutoTokenizer, AutoModelWithLMHead
```

```python
#pretrain:선행(먼저) 학습이된
tokenizer = AutoTokenizer.from_pretrained('mrm8488/t5-base-finetuned-question-generation-ap')
model = AutoModelWithLMHead.from_pretrained('mrm8488/t5-base-finetuned-question-generation-ap')
```

```python
context = "today i eat applepie."
answer = "applepie"

def get_question(answer,context):
  input_text="answer:%s context:%s </s>" %(answer,context)
  #tensor : 신경망의 단위 (신경),pt : pytorch
  #tensor : pytorch, tensorflow
  features =tokenizer([input_text],return_tensors='pt')
  print(features)
  output=model.generate(input_ids=features['input_ids'],attention_mask=features['attention_mask'])
  print(output)
  return tokenizer.decode(output[0])

get_question(answer,context)
```
